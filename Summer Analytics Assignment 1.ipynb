# Start by importing all important libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Read the csv file and assign it to a variable

```python
movies = pd.read_csv("imdb_top_1000.csv")
movies
```

# Display shape of dataframe

```python
movies.shape
```

# Print all columns of dataframe

```python
print(movies.columns.tolist())
```

# Remove the column 'Poster_Link' and update the dataframe

```python
x=movies.drop('Poster_Link', axis=1)
x
```

# Set the 'Series_Title' column as the index of dataframe

```python
df=x.set_index('Series_Title')
df
```

# Print a list of all the unique Genre

```python
df['Genre'].unique()
```

# 1. What is the Number of Directors in the given dataset ?

```python
n=len(df['Director'].unique())
n
```

# 2. What is the mean, median and standard deviation of IMDB rating in the dataset ?

```python
print("Mean :" ,df['IMDB_Rating'].mean())
print("Median :" ,df['IMDB_Rating'].median())
print("Standard Deviation :" ,df['IMDB_Rating'].std())
```

# 3. Which director has directed the most number of movies?

```python
df['Director'].value_counts().idxmax()
```

# 4. Name the Movie which has the highest Runtime ?

```python
df['Runtime'] = df['Runtime'].str.replace(r'\D', '', regex=True).astype(int)
df
df[['Runtime']].idxmax()
```

# 5.How many movies are released after year 2000 and have IMDB rating > 8.5 ?

```python
df = df.drop('Apollo 13')
df['Released_Year'] = pd.to_numeric(df['Released_Year'])
df['IMDB_Rating'] = pd.to_numeric(df['IMDB_Rating'])
movie_list = df[(df['Released_Year'] > 2000) & (df['IMDB_Rating'] > 8.5) ]
movie_list[movie_list.columns[0]].count()
```

# 6. Make a countplot of Movies directed by all directors.

```python
sns.countplot(y='Director',data=df)
sns.set(rc={'figure.figsize':(40,100)})
plt.show()
```

# 8. What type of Genre has the highest IMDB rating?

```python
print("Scatter Plot:  ")
plt.scatter(df["IMDB_Rating"], df["Genre"])
plt.show()
```

# 9. Name the top 5 movies that have earned highest Meta score.

```python
df['Series_Title'] = df.index
df.sort_values(by=['Meta_score'], ascending=False).head(5)
fig = plt.figure(figsize=(15,8))
ax = fig.add_subplot(1,1,1)
ax.bar(df["Series_Title"][:5],df["Meta_score"][:5])
plt.show()
```
